## Converting the float values to Integer vlalues 

```python
df['median_salary'] = df['median_salary'].astype(int)
df

df['average_salary'] = df['average_salary'].astype(int)
df

df['lowest_salary'] = df['lowest_salary'].astype(int)
df

df['highest_salary'] = df['highest_salary'].astype(int)
df
```

## Frequency of Countries in each Continent

```python
df['continent_name'].value_counts().plot(kind='bar', color = 'lightblue')
plt.xlabel('Continent')
plt.ylabel('Frequency')
plt.title('Frequency of Countries in Each Continent')
plt.show()
```

## Top 10 countries with Highest salaries

```python
top_10_countries = df.sort_values(by='highest_salary', ascending=False).head(10)


plt.figure(figsize=(7, 6))
plt.bar(top_10_countries['country_name'], top_10_countries['highest_salary'], color='lightgreen')
plt.xlabel('Country')
plt.ylabel('Highest Salary')
plt.title('Top 10 Countries with Highest Salaries')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

## Top 10 countries with lowest salaries

```python
bottom_10_countries = df.sort_values(by='highest_salary', ascending=True).head(10)


plt.figure(figsize=(7, 6))
plt.bar(bottom_10_countries['country_name'], top_10_countries['highest_salary'], color ='red')
plt.xlabel('Country')
plt.ylabel('Highest Salary')
plt.title('Top 10 Countries with Lowest Salaries')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

## Top 10 countries with Average salaries

```python
top_10_countries = df.sort_values(by='average_salary', ascending=False).head(10)


plt.figure(figsize=(8, 6))
plt.barh(top_10_countries['country_name'], top_10_countries['average_salary'], color='lightgreen')
plt.xlabel('Country')
plt.ylabel('average_salary')
plt.title('top 10 Countries with average Salaries')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

```

## Bottom 10 countries with Average salaries

```python
bottom_10_countries = df.sort_values(by='average_salary', ascending=True).head(10)


plt.figure(figsize=(8, 6))
plt.barh(bottom_10_countries['country_name'], bottom_10_countries['average_salary'], color= 'red')
plt.ylabel('Country')
plt.xlabel('average_salary')
plt.title('bottom 10 Countries with average Salaries')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

```
